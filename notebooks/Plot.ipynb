````python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame(
    {
        "A": [1, 5, 7, 3, 4, 0],
        "B": [9, 4, 5, 6, 1, 0],
    }
)


plt.rcParams["font.size"] = 16
plt.figure(figsize=(16, 9))
plt.xlabel("Abscissa")
plt.ylabel("Ordinates")
plt.title("My title")
plt.scatter(df["A"], df["B"], s=100)
plt.grid()
plt.savefig("plot_example.png", dpi=300)
````

I found on [Stck overflow](https://stackoverflow.com/questions/8753835/how-to-get-a-list-of-all-the-fonts-currently-available-for-matplotlib)

````python
import matplotlib.font_manager
from IPython.core.display import HTML


def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(
        font=fontname
    )


code = "\n".join(
    [
        make_html(font)
        for font in sorted(
            set([f.name for f in matplotlib.font_manager.fontManager.ttflist])
        )
    ]
)
HTML("<div style='column-count: 2;'>{}</div>".format(code))
````
